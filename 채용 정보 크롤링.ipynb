## 잡코리아 채용 정보

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import re

driver = webdriver.Chrome('C:/chromedriver/chromedriver')

driver.get("http://www.jobkorea.co.kr/recruit/joblist?menucode=search")
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[1]/dd[1]/div[1]/ul/li[3]/label/span/span').click()
driver.find_element_by_xpath('//*[@id="duty_step2_10016_ly"]/li[16]/label/span/span').click()
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[2]/dt/p').click()
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[2]/dd[2]/div[2]/dl[1]/dd/div[1]/ul/li[1]/label/span/span/em').click()
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[3]/dt/p').click()
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[3]/dd/div[1]/ul[1]/li[1]/label/span/span').click()
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[4]/dt/p').click()
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[4]/dd/div[1]/ul/li[6]/label/span/span').click()
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[6]/dt/p').click()
driver.find_element_by_xpath('//*[@id="devSearchForm"]/div[2]/div/div[1]/dl[6]/dd/div[1]/ul/li[1]/label/span/span').click()
driver.find_element_by_xpath('//*[@id="dev-btn-search"]/span').click()

company_name = []
recruit = []
contents = []
details = []
urls = []

i = 1
while 1: 
    time.sleep(1)

    element = driver.find_element_by_css_selector("body")
    text = element.get_attribute("innerHTML")
    bs = BeautifulSoup(text,"html.parser")
    
    for company in bs.select("tbody tr.devloopArea"):
        try:
            company_name.append(re.sub("㈜", "", company.select_one("td.tplCo > a").text).strip())
            recruit.append(company.select_one("td.tplTit div.titBx strong > a").text)
            contents.append([re.sub("[ ]+", " ", c.text).replace("\n", "") for c in company.select("td.tplTit div.titBx p.etc span.cell")])
            details.append(company.select_one("td.tplTit div.titBx p.dsc").text)
            urls.append("http://www.jobkorea.co.kr/" + company.select_one("td.tplTit div.titBx strong > a").attrs["href"])
        except:
            continue

    i += 1
    try:
        elem = driver.find_element_by_xpath('//*[@id="dvGIPaging"]/div/ul/li[{}]/a'.format(i)).click()
    except:
        break
print("성공")
```

```python
import pandas  as pd
df = pd.DataFrame({"회사 이름": company_name, "채용공고": recruit, "자격요건": contents, "상세내용": details, "URL" : urls})
df.to_excel('E:/채용 정보/잡코리아 기업 리스트.xlsx',index=False)
```

## 사람인 채용 정보

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import re

driver = webdriver.Chrome('C:/chromedriver/chromedriver')
driver.get("http://www.saramin.co.kr/zf_user/search?loc_mcd=101000%2C117000&cat_cd=417&exp_none=y&edu_none=y&panel_type=&search_optional_item=y&search_done=y&panel_count=y")

company_name = []
recruit = []
contents = []
details = []
urls = []


i = 1
while 1: 
    time.sleep(1)

    element = driver.find_element_by_css_selector("body")
    text = element.get_attribute("innerHTML")
    bs = BeautifulSoup(text,"html.parser")

    for company in bs.select("div.content div.item_recruit"):
        company_name.append(re.sub("\(주\)", "", company.select_one("div.area_corp strong.corp_name > a").text).strip())
        recruit.append(company.select_one("div.area_job h2.job_tit > a").text)
        contents.append([c.text for c in company.select("div.area_job div.job_condition span")])
        details.append(", ".join([j.text for j in company.select("div.area_job div.job_sector a")]))
        urls.append("http://www.saramin.co.kr/" + company.select_one("div.area_job h2.job_tit > a").attrs["href"])

    if int(bs.select("div.content_bottom a.page")[-1].text) > i:
        i += 1
        elem = driver.find_element_by_xpath('//*[@id="recruit_info_list"]/div[2]/div/a'.format(i)).click()
    else:
        break
print("성공")
```

```python
import pandas  as pd
df = pd.DataFrame({"회사 이름": company_name, "채용공고": recruit, "자격요건": contents, "상세내용": details, "URL" : urls})
df.to_excel('E:/채용 정보/사람인 기업 리스트.xlsx',index=False)
```

## 로켓펀치 채용 정보

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
import re

driver = webdriver.Chrome('C:/chromedriver/chromedriver')
driver.get("https://www.rocketpunch.com/jobs?career_type=1&job=1&location=%EC%84%9C%EC%9A%B8%ED%8A%B9%EB%B3%84%EC%8B%9C&page=1&specialty=Python")

company_name = []
recruit = []
contents = []
details = []
urls = []

i = 1
while 1: 
    time.sleep(1)

    element = driver.find_element_by_css_selector("body")
    text = element.get_attribute("innerHTML")
    bs = BeautifulSoup(text,"html.parser")

    for company in bs.select("div#company-list div.company"):
        company_name.append(company.select_one("div.company-name a h4.name strong").text)
        recruit.append([re.sub("[ ]+", " ", c.text.replace("\n", " ").strip()) for c in company.select("div.job-detail div")])
        contents.append(company.select_one("div.nowrap").text.replace("\n", "").strip())
        details.append(company.select_one("div.description").text.replace("\xa0", ""))
        urls.append("https://www.rocketpunch.com" + company.select_one("div.company-name a").attrs["href"])

    if i == 1:
        elem = driver.find_element_by_xpath('//*[@id="search-results"]/div[3]/div/a[2]'.format(i)).click()
    elif i == 5:
        break
    else:
        elem = driver.find_element_by_xpath('//*[@id="search-results"]/div[2]/div/a[2]'.format(i)).click()
        
    i += 1
    time.sleep(1)
    
    try:
        driver.find_element_by_xpath('//*[@id="fb-like-modal"]/div[1]/i').click()
    except:
        continue
print("성공")
```

```python
import pandas  as pd
df = pd.DataFrame({"회사 이름": company_name, "채용공고": recruit, "자격요건": contents, "상세내용": details, "URL" : urls})
df.to_excel('E:/채용 정보/로켓펀치 기업 리스트.xlsx',index=False)
```

```python
import pandas  as pd
xlsx = pd.read_excel('E:/채용 정보/로켓펀치 기업 리스트.xlsx')
xlsx
```
